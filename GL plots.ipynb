```python
x1=np.arange(1,5,1)

fig1 = plt.figure(figsize=(15,10))
# gl472c1=np.array([0.287,0.177,0.204,0.2435])
# gl472c2=np.array([0.287,0.177,0.207,0.2435])
gl573c1=np.array([0.167,0.131,0.108,0.057])
gl573c2=np.array([0.167,0.131,0.108,0.057])

# plt.plot(x1,gl472c1,'bo-')
# plt.plot(x1,gl472c2,'ro-')
plt.plot(x1,gl573c1,'bo-')
plt.plot(x1,gl573c2,'ro-')

plt.xlabel("Position")
plt.ylabel("Thickness of Glue Layer (mm)")
# plt.title("Measurements of Glue Layer of Module 472")
plt.title("Measurements of Glue Layer of Module 573")

plt.legend(['Case 1','Case 2'])
plt.show()
```
